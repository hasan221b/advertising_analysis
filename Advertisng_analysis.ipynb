```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('advertising.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.duplicated().sum()
```

The datasetdoes not contain any duplicated values or null values, indicating that each entry is unique and complete.

## Time spent

```python
data.groupby('Clicked on Ad')['Daily Time Spent on Site'].mean()
```

An interesting finding is that the average time spent on the site by individuals who did not click on the ad is higher compared to those who did click on the ad, with a difference of 44.6%. This suggests that the amount of time spent on the site does not appear to significantly influence the decision to click on the ad.

## Age

```python
plt.figure(figsize=(10, 4))  # Adjust the figsize according to your preference
plt.subplot(1, 2, 1)  # Create a subplot for the histogram
plt.hist(data['Age'], bins='auto',edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
```

### Internet Usage

```python
data.groupby('Clicked on Ad')['Daily Internet Usage'].mean()
```

The analysis reveals that the number of non-clickers exceeds the number of clickers by a significant margin of 32.10%.

## Heatmap

```python
newdf = data[['Age','Daily Time Spent on Site','Area Income','Daily Internet Usage','Clicked on Ad']]
correlation_matrix = newdf.corr()

# Plotting the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Upon analyzing the provided heatmap and the analysis above several observations can be made:

1. The correlation between clicking on an ad and the other features appears to be weak, with the exception of Age.
2. It can be concluded that the age of the user has the most significant impact on ad clicks.
3. The age range of 35 to 50 witnessed the highest number of clicks.
4. Specifically, the ages of 45, 38, and 36 stand out as the highest among those who clicked on ads.
